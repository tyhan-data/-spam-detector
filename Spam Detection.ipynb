# Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report, confusion_matrix
from joblib import dump
```

# Load Dataset

```python
df = pd.read_csv(r"C:\Dataset for practice\spam.csv", encoding='latin-1')
```

```python
df.head()
```

# Drop Unwanted Columns

```python
df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'], inplace=True)
```

```python
df.head(3)
```

# Rename Columns Name

```python
df.rename(columns={'v1':'Target', "v2":"Text"}, inplace=True)
```

```python
df.head(2)
```

# Removing Duplicates Data

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

# Feature and Target Selection

```python
x = df['Text']
y = df['Target']
```

# Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

# CountVectorizer Section

```python
cv = CountVectorizer(stop_words = 'english')

x_train_cv = cv.fit_transform(x_train)
x_test_cv = cv.transform(x_test)
```

```python
x_test_cv.shape
```

# Naive Bayes Model

```python
model = MultinomialNB()
model.fit(x_train_cv, y_train)
```

# Prediction

```python
y_pred = model.predict(x_test_cv)
```

# Evaluation

```python
print("Classification Report:\n",classification_report(y_test, y_pred))
```

```python
print("Confusion Matrix:\n",confusion_matrix(y_test, y_pred))
```

# Simple Test

```python
text = cv.transform(['hello sir ! you won 1000 dollar.'])
output = model.predict(text)
print(output)
```

# Custom Test

```python
def prediction(msg):
    input_msg = cv.transform([msg])
    result = model.predict(input_msg)
    return result

msg = ("Hello bro! How are you.")
print("Message:",msg)
print('Prediction:',prediction(msg),'\n')

msg2 = ('you won 1000 dollar.')
print("Message:",msg2)
print('Prediction:',prediction(msg2))
```

# Saving Model

```python
print(dump(model, 'Spam_detector.joblib'))
print(dump(cv, 'Count_Vectorizer.joblib'))
```
